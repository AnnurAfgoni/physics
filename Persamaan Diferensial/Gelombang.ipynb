# Theory

Bentuk Umum :

$$\gamma = A \sin (\omega t \pm kx)$$

* (-) negatif jika gelombang merambat ke kanan
* (+) positif jika gelombang merambat ke kiri

* cepat rambat gelombang $\omega = \lambda f$ atau $\omega = \frac{\lambda}{T}$

```python
"""
Program untuk membuktikan arah gerak rambat gelombang
yang menetukan arah rambat adalah tanda dari bilangan gelombang (k)

ubahlah nilai omega dan k supaya bisa melihat perbedaan pada animasinya
"""

# bilangan gelombang
k = 1

# cepat rambat gelombang
omega = 10

# data input
x = np.linspace(0, 10, 1000)
y = []

plt.figure(figsize=(20, 8))
for i in range(10):
    t = i
    
    # ubah tanda minus menjadi plus dan sebaliknya
    # untuk melihat arah gerak gelombang
    yi = np.sin(omega*t + k*x) 
    y.append(list(yi))
    
    # plot
    plt.plot(x, yi, label = "t = "+ str(t))

plt.legend()
plt.show()
```

```python
import imageio
from IPython.display import Image

# membuat gambar
for n in range(len(y)):
    plt.plot(x, y[n])
    plt.axis("off")
    plt.xlim([0,10])
    filename = "gerak gelombang" + str(n) + ".png"
    plt.xticks(ticks=None)
    plt.savefig(filename)
    plt.close("all")
    
# membuat animasi gif
images = []
for i in range(len(y)):
    filename = "gerak gelombang" + str(i) + ".png"
    images.append(imageio.imread(filename))
imageio.mimsave("gerak_gelombang.gif", images)

# show animasi
Image("gerak_gelombang.gif", width=500, height=300)
```

**Sudut Fase Gelombang**
adalah sudut yang telah ditempuh benda ketika mencapai suatu titik.

$$\theta_p = (\omega t - kx) = 2\pi(\frac{t}{T}-\frac{x}{\lambda})$$

ketika waktu awal ($t = 0$) yang berarti $\theta_p = \frac{2\pi x}{\lambda}$. Ketika kita menuliskan persamaan $\sin(\frac{2\pi x}{\lambda})$ berarti itu merupakan persamaan gelombang awal ketika $t = 0$
