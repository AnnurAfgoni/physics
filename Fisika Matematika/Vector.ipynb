# Perkaian Skalar Vector

1. Asosiatif: $(\alpha\beta) x = \alpha(\beta x)$
2. Left Distributif: $(\alpha + \beta)x = \alpha x + \beta x$
3. Right Distributif: $X(\alpha + \beta) = x\alpha + x\beta$
4. Right distributif for vector addition: $\alpha(x+y)=\alpha x + \alpha y$

# Plot Vector

`plt.quiver()`

```python
# origin 
x = [0]
y = [0]

# direction
u = [2]
v = [1]

# plot
plt.quiver(x, y, u, v, color='b', units='xy', scale=1)
plt.title('Single Vector')

# x-lim and y-lim
plt.xlim(-2, 5)
plt.ylim(-2, 2.5)

plt.grid()
plt.show()
```

# Ortogonal Vector

**Ortogonal Vector** : Two vectors are orthogonal to each other when their dot product is 0.

definition:

$$A \cdot B = \sum_{i=1}^{n}a_{i}b_{i}=A^{T}B=0$$

**Question** are $v_1 = [1, -2, 4]$ and $v_2 = [2, 5, 5]$ Orthogonal ?

**Answers:**

```python
import numpy as np

# vectors
v1 = np.array([[1, -2, 4]])
v2 = np.array([[2, 5, 2]])

result = np.dot(v1, v2.T)

if result == 0:
    print(" yes they are orthogonal")
else:
    print("no, they don't")
```

# Unit Vector

definition of unit vector:

$$\hat{a} = \frac{A}{|A|}$$

```python
# define a vector

A = np.array([3,4])
magnitude = np.sqrt(np.sum(A**2))
print(A/magnitude)
```

```python
# plot to see the unit vector

plt.plot([0, 3], [0, 4], "ro-")
plt.plot([0, 0.6], [0, 0.8], "go-")
plt.grid()
plt.plot()
```

# Orthonormal Vectors

these vectors are orthogonal to each other and individually they also have unit magnitude.

```python
# vectors
v1 = np.array([[1, -2, 4]])
v2 = np.array([[2, 5, 2]])

v1_norm = np.sqrt(np.sum(v1**2))
v2_norm = np.sqrt(np.sum(v2**2))

v1_unit = v1/v1_norm
v2_unit = v2/v2_norm

print(v1_norm)
print(v2_unit)
```

# Projection

```python
# vector 1 terhadap vector 2
vec_1 = np.array([3, 4])
vec_2 = np.array([1, 1])

vec2_norm = np.sqrt(np.sum(vec_2**2))

projection = (np.dot(vec_1, vec_2)/vec2_norm**2)*vec_2

print(projection)
```

```python
plt.plot([0, vec_1[0]], [0, vec_1[1]], "ro-")
plt.plot([0, projection[0]], [0, projection[1]], "go-")
plt.grid()
plt.show()
```

```python
np.sqrt(np.sum(vec_1**2))
```

```python
np.sqrt(np.sum(projection**2))
```

```python
# vector 2 terhadap vector 1
vec_1 = np.array([3, 4])
vec_2 = np.array([1, 1])

vec1_norm = np.sqrt(np.sum(vec_1**2))

projection = (np.dot(vec_2, vec_1)/vec1_norm**2)*vec_1

print(projection)
```

```python
plt.plot([0, vec_2[0]], [0, vec_2[1]], "ro-")
plt.plot([0, projection[0]], [0, projection[1]], "go-")
plt.grid()
plt.show()
```

```python
np.sqrt(np.sum(vec_2**2))
```

```python
np.sqrt(np.sum(projection**2))
```
